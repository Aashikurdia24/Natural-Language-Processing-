## **Text Classification in NLP**

* Text classification is a fundamental task in natural language processing (NLP) that involves assigning predefined categories or labels to text documents.

* This process enables the automated sorting and organization of textual data, facilitating the extraction of valuable information and insights from large volumes of text.


* Text classification is widely used in various applications, including sentiment analysis, spam detection, healthcare, fake new detection, customer support, language detection, topic labelling, and document categorization.

## **There are 3 types of text classification :**

1. Binary classification : only 2 classes are present such as (yes ,no) or (1,0) Email spam classifier

2. Multi class classification : here multi classes are present such as (product A ,B and C) Reading a news article and then classifying into its class based on genre

3. Multi label classification : There can be multiple outputs for a single classification. Like a single news article can be classified into different classes. For example : A bollywood news article can be classified into classes like glamour, trending, youth, celebrity news etc

## **Steps for implementing text classification :**

1.  Data acquisition: Getting a dataset from API, or collecting it from various platforms

2. Text pre processing : Cleaning the textual data like doing tokenization, stemming, removing stop words, emojis, html tags, punctuations, links, correcting the spelling mistakes, special characters etc depending on the applications requirements

3. Text vectorization : Converting text into numbers and we can do it so by using techniques like BoW, TF-IDF, Word2Vec

4. Creating a model : Use either ML or DL approach. For ML approach we can use techniques like Naive Bayes, Random Forest and for DL approach we can use techniques like RNNs, LSTM, BERT

5. Performing the evaluation of the model by : using metrics like accuracy score, precision and recall, confusion matrix, ROC AUC Curve

6. Deploying the pipeline

## **1.Term Frequency-Inverse Document Frequency tf-idf**


**Term Frequency** : measures the frequency of a term within a document.

It is calculated as the ratio of the number of times a term occurs in a document to the total number of terms in that document.


**Inverse Document Frequency** : IDF measures the rarity of a term across a collection of documents.

It is calculated as the logarithm of the ratio of the total number of documents to the number of documents containing the term.

**TF-IDF** : The TF-IDF score for a term in a document is obtained by multiplying its TF and IDF scores.

## **2. Bag of Words**

* BoW represent text data by counting the occurrence of the words in a document.

* It disregards grammar and word order, treating each each document as a collecion of words.

* BoW doesn't consider the importance of words in the document or across the corpus ; it only counts their occurrences

## **3. Word2Vec**

* Word2Vec is a word embedding technique in NLP that represents words as vectors in a continuous space.

* The underlying assumption of Word2Vec : when two words share similar context also share a similar meaning and consequently a similar (vector)representation from the model.

## **Advantages of Word2Vec**
1. Understands semantic relationship between words
2. low dimensional vectors between [100-300] this allows to compute fast
3. dense vector : there are non-zero vectore which helps to reduce overfitting

## **Continuous Bag of Words [CBoW]**

CBoW predicts the target-word based on its surrounding words

## **Skip-gram**

skip gram works in the exact opposite way to CBoW

Here ,we take an input word and except the model to predict its surrounding words

sentence : "I like apple and mango."

X[input]--------->Y[predict surrounding words]

like -------> I,apple

and --------> apple ,mango
